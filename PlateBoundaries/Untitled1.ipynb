```python
import numpy as np

points = np.array([[0, 0], [0, .4], [1, 0], [1, 1], [0.2, 0.2], [0.5,0.7]])
from scipy.spatial import Delaunay
tri = Delaunay(points)
```

```python
import matplotlib.pyplot as plt
plt.triplot(points[:,0], points[:,1], tri.simplices)
plt.plot(points[:,0], points[:,1], 'o')
plt.show()
```
